# Test Data Processing

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os


data_path = '../src/results/'
data_file = 'random_moves_results_20180713162316279796.csv'
data = pd.read_csv(os.path.join(data_path, data_file))
data.head()
```

```python
vals, cnts = np.unique(data.loc[:, 'max_tile'], return_counts=True)
plt.bar(vals, cnts, width=8, tick_label=vals)
plt.xlabel('Maximum tile value')
plt.ylabel('Number of occurrences')
```

```python
plt.hist(data.loc[:, 'score'], 40)
plt.xlabel('Score')
plt.ylabel('Number of occurences')
```

```python
plt.plot(data.loc[:, 'score'], '.')
plt.xlabel('Iteration')
plt.ylabel('Score')
```
